# NYC Film Permits — Most Filmed Blocks

```python
import pandas as pd

# Load the data
df = pd.read_excel('Film_Permits_20260416.xlsx')
print('Total permits:', len(df))
df.head(3)
```

```python
# Preview the raw ParkingHeld column
df['ParkingHeld'].head(3)
```

```python
# Each cell contains multiple blocks separated by commas.
# Step 1: split each cell into a list of blocks
df['blocks'] = df['ParkingHeld'].str.split(',')

# Step 2: explode — turns each block into its own row
df_blocks = df.explode('blocks')

# Step 3: clean up whitespace and make uppercase so counts are consistent
df_blocks['blocks'] = df_blocks['blocks'].str.strip().str.upper()

# Drop any empty rows that appeared from trailing commas
df_blocks = df_blocks.dropna(subset=['blocks'])
df_blocks = df_blocks[df_blocks['blocks'] != '']

print('Total block-permit appearances:', len(df_blocks))
df_blocks['blocks'].head(5)
```

```python
# Count how often each block appears and show the top 20
top_blocks = df_blocks['blocks'].value_counts().head(20)
top_blocks
```

```python
# Same thing but as a nicer table with a rank column
top_blocks_df = top_blocks.reset_index()
top_blocks_df.columns = ['Block', 'Permit Count']
top_blocks_df.index = top_blocks_df.index + 1  # start ranking at 1
top_blocks_df
```

```python
# Bonus: top 5 blocks per borough
top_per_borough = (
    df_blocks.groupby('Borough')['blocks']
    .value_counts()
    .groupby(level=0)
    .head(5)
)
top_per_borough
```
